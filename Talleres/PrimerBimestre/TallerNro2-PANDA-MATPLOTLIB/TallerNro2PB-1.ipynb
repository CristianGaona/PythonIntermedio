## Importar ambas tablas de datos en python usando pandas. Poner la columna del índice en "MESS_DATUM" y analizar los valores de la columna como fechas.

```python
garmisch  = pd.read_csv('data/garmisch.txt', sep=";")

# Bucle que permite cambiar la por tip Date
for i in garmisch['MESS_DATUM']:
    s = str(i)
    date = datetime(year=int(s[0:4]), month=int(s[4:6]), day=int(s[6:8]))
    garmisch['MESS_DATUM'] =  garmisch['MESS_DATUM'].replace(i,date)

# Para cambiar de id del dataframe
garmisch_index = pd.DataFrame(garmisch.set_index('MESS_DATUM'))
garmisch_index

#print(garmisch.dtypes)
```

```python
zugspitze = pd.read_csv('data/zugspitze.txt', sep=";")
# Bucle que permite cambiar la por tip Date
for i in zugspitze['MESS_DATUM']:
    s = str(i)
    date = datetime(year=int(s[0:4]), month=int(s[4:6]), day=int(s[6:8]))
    zugspitze['MESS_DATUM'] =  zugspitze['MESS_DATUM'].replace(i,date)


# Para cambiar de id del dataframe
zugspitze_index = pd.DataFrame(zugspitze.set_index('MESS_DATUM'))
zugspitze_index

#print(zugspitze.dtypes)
```
