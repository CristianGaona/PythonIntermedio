````python
df_pais=earthquake["place"].str.split(", ", n=1, expand = True)
df_pais.columns = ['Place', 'Country']
df_pais['Country']
#earthquake = pd.concat([earthquake, df_pais["Country"]], axis=1)
earthquake['Country'] = df_pais['Country']
earthquake
````

````python
import us
estados=us.states.STATES
#df_remplazo = pd.DataFrame(earthquake["Country"]) 
for i in range(len(estados)):
  auxEstados=str(estados[i])
  earthquake['Country']= earthquake['Country'].replace(auxEstados, "EEUU")
earthquake.head(10)
````
